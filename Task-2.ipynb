**Data Loading and Initial Exploration**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('titanic.csv')
```

```python
# Display the first few rows
print(df.head())

# Get information about data types and missing values
print(df.info())

# Summary statistics
print(df.describe(include='all'))
```

**Handling Missing Values**

```python
# Check missing values
print(df.isnull().sum())

# Handling missing values
# Fill missing 'Age' values with the median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Drop rows with missing 'Embarked' values
df.dropna(subset=['Embarked'], inplace=True)
```

**Convert categorical variables to appropriate formats**

```python
# Convert categorical variables to appropriate formats
df['Sex'] = df['Sex'].astype('category').cat.codes
df['Embarked'] = df['Embarked'].astype('category').cat.codes
print(df)                                                                  
```

**Visualizations**

**Age Distribution**

```python
# Distribution of ages using hist
plt.figure(figsize=(10, 6))
plt.hist(df['Age'].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**Survival Count**

```python
# Survival rate using countplot
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=df , palette={0: 'magenta', 1: 'cyan'})
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

**Survival Rate by Gender**

```python
# Survival rate by gender using barplot
sns.barplot(x='Sex', y='Survived', data=df , palette={0: 'blue', 1: 'pink'})
plt.title('Survival Rate by Gender')
plt.xlabel('Gender (0 = Male, 1 = Female)')
plt.ylabel('Survival Rate')
plt.show()
```

**Survival Rate by Passenger Class**

```python
# Survival rate by Passenger Class using barplot
sns.barplot(x='Pclass', y='Survived', data=df , palette={1: 'purple', 2: 'red', 3: 'green',})
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```
